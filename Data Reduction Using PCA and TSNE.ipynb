# MNIST Using PCA and TSNE

```python
#loading MNIST
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
d=pd.read_csv('mnist_train.csv')
```

```python
l=d['label']
```

```python
d1=d.drop('label',axis=1)
```

```python
#display
plt.figure(figsize=(7,7))
idx=100
grid_data=d1.iloc[idx].as_matrix().reshape(28,28)
plt.imshow(grid_data,interpolation='none',cmap='gray')
plt.show()
print(l[idx])
```

# PCA for Visualization

```python
labels=l.head(40000)
data=d1.head(40000)
```

```python
print(data.shape)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
stand_data=StandardScaler().fit_transform(data)
```

```python
from sklearn import decomposition
```

```python
pca=decomposition.PCA()
```

```python
pca.n_components=2
```

```python
pca_data=pca.fit_transform(stand_data)
```

```python
#appending label
new=np.vstack((pca_data.T,labels)).T
```

```python
#new Dataframe
df=pd.DataFrame(data=new,columns=('v1','v2','label'))
```

```python
import seaborn as se
```

```python
se.FacetGrid(df,hue='label',size=6)\
  .map(plt.scatter,'v1','v2')\
  .add_legend()
plt.show()
```

# PCA for reduction

```python
pca.n_components=784
pca_data=pca.fit_transform(stand_data)
percentage_var_exp=pca.explained_variance_/np.sum(pca.explained_variance_)
```

```python
cum_var=np.cumsum(percentage_var_exp)
```

```python
#plotting
plt.figure(1,figsize=(6,4))
plt.clf()
plt.plot(cum_var,linewidth=2)
plt.axis('tight')
plt.grid()
plt.xlabel('features')
plt.ylabel('cum variance')
plt.show()
```

# TSNE

```python
from sklearn.manifold import TSNE 
```

```python
data_1=stand_data[0:1000,:]
label_1=labels[0:1000]
```

```python
model=TSNE(n_components=2,random_state=0,perplexity=100,n_iter=5000)
```

```python
tsne_data=model.fit_transform(data_1)
```

```python
tsne_data=np.vstack((tsne_data.T,label_1)).T
tsne_df=pd.DataFrame(data=tsne_data,columns=('dim1','dim2','label'))
#plotting
se.FacetGrid(tsne_df,hue='label',size=6)\
  .map(plt.scatter,'dim1','dim2')\
  .add_legend()
plt.show()
```

```python
#changng perplexity
model=TSNE(n_components=2,random_state=0,perplexity=50,n_iter=5000)
tsne_data=model.fit_transform(data_1)
tsne_data=np.vstack((tsne_data.T,label_1)).T
tsne_df=pd.DataFrame(data=tsne_data,columns=('dim1','dim2','label'))
#plotting
se.FacetGrid(tsne_df,hue='label',size=6)\
  .map(plt.scatter,'dim1','dim2')\
  .add_legend()
plt.show()
```

# PCA on Cancer dataset

```python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
d=pd.read_csv('cancer.csv')
```

```python
d1=d.drop('diagnosis',axis=1)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
from sklearn import decomposition
```
